```python
import pandas as pd
```

```python
mydict = {
    "developer_id": ["001", "002", "003"],
    "name": ["Thomas", "Andrew", "John"],
    "age": [36, 21, 44],
    "country": ["Germany", "United States", "Scotland"],
    "city": ["Berlin", "Boston", "Edinburgh"],
}

df_developers = pd.DataFrame(mydict)
df_developers.head()
```

```python
mydict = {
    "developer_id": ["001", "002", "002", "002"],
    "skill": ["Python", "Java", "JavaScript", "React"],
    "level": ["excellent", "very good", "good", "good"],
}

df_skills = pd.DataFrame(mydict)
df_skills.head()
```

```python
df_merged = pd.merge(df_developers, df_skills, how="inner", on="developer_id")
df_merged
```
